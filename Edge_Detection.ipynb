### OpenCV Edge Detection Using Canny Algorithm

```python
import cv2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def image_detection(image):
    edges = cv2.Canny(image, 200, 100)
    plt.figure(figsize=(8,5))
    plt.subplot(121)
    plt.axis('off')
    plt.title('original')
    plt.imshow(image[:,:,[2,1,0]])
    plt.subplot(122)
    plt.axis('off')
    plt.title('edges')
    plt.imshow(edges, cmap='gray')
    plt.tight_layout()
    plt.show()
```

```python
image = cv2.imread("C:/Users/User/Pictures/sho/Picture_A.png")
image_detection(image)
```
